```python
# titanic = pd.read_csv('https://raw.githubusercontent.com/mrBronnWow/Curso_Beginners/main/1_Dataset_titanic/train.csv')
# titanic.to_csv("titanic.csv", index=False)
# Info https://www.kaggle.com/c/titanic/overview
titanic = pd.read_csv("titanic.csv")
titanic
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
#Si Age es vacio es mas probable que sean de clase baja
#Si Cabin es vacÃ­o es que no tenian camarote
#Si Embarked es vacio es que no dice en quÃ© puerto embarcaron. Gente cheta. SobreviviÃ³
#Reemplazo valores legibles
titanic["Embarked"] = titanic["Embarked"].replace({"C": "Cherbourg", "Q": "Queenstown", "S": "Southampton"})
titanic["Cabin"]=titanic["Cabin"].fillna("No Cabin").apply(lambda x: "Some cabin" if x != "No Cabin" else x)


for nancol in [col for col in titanic.columns if titanic[col].hasnans]:
    print(titanic[nancol].isna().value_counts())

```

```python
titanic[titanic["Age"].isna()]
```

```python
titanic[titanic["Age"].isna()].describe()
```

```python
#Relacion de carentes de Age por clase
titanic[titanic["Age"].isna()]["Pclass"].value_counts() / titanic["Pclass"].value_counts()
```

```python
#Relacion de carentes de Age por sexo
titanic[titanic["Age"].isna()]["Sex"].value_counts() / titanic["Sex"].value_counts()
```

```python
#Relacion de carentes de Age por survived
titanic[titanic["Age"].isna()]["Survived"].value_counts() / titanic["Survived"].value_counts()
```

```python
#Si tenian cabina 
print(titanic.groupby("Cabin")["Survived"].sum())
print(titanic.groupby("Cabin")["PassengerId"].count())
titanic.groupby("Cabin")["Survived"].sum() / titanic.groupby("Cabin")["PassengerId"].count()
```

```python
#Debe ser un error de carga
titanic[titanic["Embarked"].isna()]
```

```python
#Relacion de supervivencia por sexo
print(titanic.groupby("Sex")["Survived"].sum())
print(titanic.groupby("Sex")["PassengerId"].count())
titanic.groupby("Sex")["Survived"].sum()/titanic.groupby("Sex")["PassengerId"].count()
```

```python
#Relacion de supervivencia por clase

print(titanic.groupby("Pclass")["Survived"].sum())
print(titanic.groupby("Pclass")["PassengerId"].count())
titanic.groupby("Pclass")["Survived"].sum()/titanic.groupby("Pclass")["PassengerId"].count()
```

```python
#Relacion de supervivencia por puerto de salida

print(titanic.groupby("Embarked")["Survived"].sum())
print(titanic.groupby("Embarked")["PassengerId"].count())
titanic.groupby("Embarked")["Survived"].sum()/titanic.groupby("Embarked")["PassengerId"].count()
```

```python
#Relacion de supervivencia por clase

print(titanic.groupby(["Pclass","Sex"])["Survived"].sum())
print(titanic.groupby(["Pclass","Sex"])["PassengerId"].count())
titanic.groupby(["Pclass","Sex"])["Survived"].sum()/titanic.groupby(["Pclass","Sex"])["PassengerId"].count()
```
