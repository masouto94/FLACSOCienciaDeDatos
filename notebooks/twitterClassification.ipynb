```python
offensive = df_tw[df_tw["does_this_tweet_contain_hate_speech"] == "The tweet contains hate speech"]["tweet_text"]
#Remove RTs and tags
offensive_no_rt = offensive[~offensive.str.startswith("RT")]
offensive_no_rt = offensive_no_rt.apply(lambda tw: re.sub("(\@\w+)","",tw)).apply(str.strip)
offensive_no_rt
```

```python
offensive_tokenized = offensive_no_rt.apply(str.split)
offensive_tokenized
```

```python
def filter_symbols(array:list) -> list:
    pattern = re.compile("\w")
    filtered = [token for token in array if len(re.findall(pattern, token)) > 0]
    without_symbols = [re.sub("\W","",token) for token in filtered]
    return without_symbols
offensive_clean = offensive_tokenized.apply(filter_symbols)
offensive_clean
```

```python
offensive_words = offensive_clean.explode().apply(str.lower)
offensive_words = offensive_words.reset_index().drop("index", axis=1)
offensive_words
```

```python
stops = set(stopwords.words("english"))
offensive_true_words = offensive_words[~offensive_words.isin(stops)]
offensive_true_words.value_counts()[0:20]
```
